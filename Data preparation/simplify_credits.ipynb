```python
#-------------------------------------------------------------------------------------------------
# Semplifica il file credits.csv, eliminando le colonne inutili e prelevando parte delle informazioni su attori e crew
#-------------------------------------------------------------------------------------------------
import pandas as pd
import csv
import ast

credits = pd.read_csv('credits_originale.csv')
crediti_attori = []
nome_regista = ''
nome_sceneggiatore = ''
nuovicrediti = pd.DataFrame(columns=['id', 'attore1', 'attore2', 'attore3', 'regista', 'sceneggiatore'])

#scorre il dataframe e per ogni riga prende i primi tre nomi di attori e li mette in una lista
#ciclo delle righe della tabella credits
with open('credits_simple.csv', 'a', newline='') as csvfile:
    
    writer = csv.DictWriter(csvfile, fieldnames=['id', 'attore1', 'attore2', 'attore3', 'regista', 'sceneggiatore'])
    if csvfile.tell() == 0:
        writer.writeheader()
    
    for i in range(len(credits)):
            attori = ast.literal_eval(credits['cast'][i])
            crew = ast.literal_eval(credits['crew'][i])
            crediti_attori.clear()
            directorfound = False
            screenplayfound = False

            #cicli interni alle celle delle colonne di cast e crew
            for d in attori[:3]:  
                    crediti_attori.append(d['name'])
            for d in crew: 
                    if d['job'] == 'Director' and directorfound == False:
                            nome_regista = d['name']
                            directorfound = True

                    elif d['job'] == 'Screenplay' and screenplayfound == False:
                            nome_sceneggiatore = d['name']
                            screenplayfound = True

            #controlla se la posizione i nella lista credits non esiste e se non esiste la crea e la inizializza con ''
            while len(crediti_attori) < 3:
                    crediti_attori.append('')

            var_id = credits['id'][i].astype(int)
            var_attore1 = crediti_attori[0]
            var_attore2 = crediti_attori[1]
            var_attore3 = crediti_attori[2]

            #Il dizionario corrisponde alla nuova riga da inserire
            dizionario_credits = {
                'id': var_id,
                'attore1': var_attore1,
                'attore2': var_attore2,
                'attore3': var_attore3,
                'regista': nome_regista,
                'sceneggiatore': nome_sceneggiatore
            }

            writer.writerow(dizionario_credits)
 
```
